# Train test split
Always evaluate your final model on a set that has not been used in training

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test=train_test_split(df,test_size=0.25)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
df_train.head()
```

# Data Modelling
Splitting data into numpy arrays

```python
x_train=df_train.drop(columns=["target","target_name"]).values
y_train=df_train["target"].values
```

```python
y_train    
```

# Modelling
Should beat atleast 33.33%

```python
def single_feature_prediction(petal_length):
    """Predicts the iris species given petal length"""
    if petal_length<2.5:
        return 0
    elif petal_length<4.8:
        return 1
    else:
        return 2
```

```python
x_train[:,2]
```

```python
manual_y_prediction=np.array([single_feature_prediction(val) for val in x_train[:, 2]])
```

```python
manual_model_accuracy = np.mean(manual_y_prediction==y_train)
```

```python
print("Manual model accuracy : ",manual_model_accuracy*100)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
#xt stands for "x train" and xv for "x validation"
xt,xv,yt,yv = train_test_split(x_train,y_train,test_size=0.25)
```

```python
model.fit(xt,yt)
```

```python
y_pred=model.predict(xv)
```

```python
np.mean(y_pred==yv)
```

```python
model.score(xv,yv) #Dont evaluate your model on the same data that was used for training
```

### Using Cross Validation

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model,x_train,y_train,cv=5,scoring="accuracy")
```

```python
np.mean(accuracies)
```

```python
y_pred = cross_val_predict(model,x_train,y_train,cv=5)
```

```python
predicted_correctly_mask = y_pred == y_train
```

```python
predicted_correctly_mask
```

```python
not_predicted_correctly = ~predicted_correctly_mask
```

```python
not_predicted_correctly
```

```python
x_train[not_predicted_correctly]
```

```python
df_predictions = df_train.copy()
```

```python
df_predictions["correct prediction"] = predicted_correctly_mask
```

```python
df_predictions["prediction"] = y_pred
```

```python
df_predictions["prediction_label"] = df_predictions["prediction"].map({0:"setosa",1:"versicolor",2:"virginica"})
```

```python
df_predictions.head()
```

```python
sns.scatterplot(x="petal length (cm)",y="petal width (cm)",hue="prediction_label",data=df_predictions)
```

```python
sns.scatterplot(x="petal length (cm)",y="petal width (cm)",hue="target_name",data=df_predictions)
```

```python
def plot_incorrect_predictions(df_predictions,x_axis_feature,y_axis_feature):
    fig,axs = plt.subplots(2,2,figsize = (10,10))
    axs=axs.flatten()
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="prediction_label",data=df_predictions,ax=axs[0])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="target_name",data=df_predictions,ax=axs[1])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="correct prediction",data=df_predictions,ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
```

```python
plot_incorrect_predictions(df_predictions,"petal length (cm)","petal width (cm)")
```
